```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize = (9,4))

sns.histplot(data = data ,x = 'Year', color = 'blue',
             edgecolor = 'linen', alpha = 0.7, bins = 5)

plt.title("Histogram of Year")
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize = (9,4))

sns.histplot(data = data ,x = 'Year', color = 'blue',
             edgecolor = 'linen', alpha = 0.5, bins = 5)

plt.title("Histogram of Year")
plt.xlabel('Year')
plt.ylabel('Count')
plt.vlines(data['Year'].mean(), ymin = 0, ymax = 4000, colors='blue', label='Mean')
plt.vlines(data['Year'].median(), ymin = 0, ymax = 4000, colors='red', label='Median')
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (9,4))

sns.histplot(data = data ,x = 'Date', color = 'green',
             edgecolor = 'linen', alpha = 0.7, bins = 5)

plt.title("Histogram of Date")
plt.xlabel('Date')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize = (9,4))

sns.distplot(data['Year'], color = 'blue')

plt.title("Distplot of Year")
plt.xlabel('Year')
plt.ylabel('Density')
plt.show()
```
